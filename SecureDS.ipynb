```python
import pandas as pd
import numpy as np
import hashlib as hl
import utils
from os import path
opt=input("choose an option:\n1)Hash anonymizer\n2)k anonymizer\n")
location=input("Enter the input file path:\n")
assert path.exists(location), "File not found"
olocation=input("Enter the output file path:\n")
if(opt=='1'):
    ds=pd.read_csv(location)
    op=pd.read_csv(location)
    display(ds)
    i=0
    op['PIN']=op['PIN'].astype(str)
    for X in ds['PIN']:
        print(X)
        s1=hl.sha1()
        s1.update(str(X).encode('utf-8'))
        op.at[i,'PIN']=s1.hexdigest()
        i+=1
    display(op)
    i=0
    for Y in ds['PASS']:
        print(Y)
        s1=hl.sha1()
        s1.update((Y).encode('utf-8'))
        op.at[i,'PASS']=s1.hexdigest()
        i+=1
    display(op)
    op.to_csv(olocation)
elif(opt=='2'):
    ds=pd.read_csv(location)
    op=pd.read_csv(location)
    display(ds)
    i=0
    op['PIN']=op['PIN'].astype(str)
    for X in ds['PIN']:
        print(X)
        op.at[i,'PIN']='*'
        i+=1
    display(op)
    i=0
    for Y in ds['PASS']:
        print(Y)
        op.at[i,'PASS']='*'
        i+=1
    display(op)
    op.to_csv(olocation)
```
